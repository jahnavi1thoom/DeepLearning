# **Diabetes Risk Prediction(Binary Classification)**

Neural Network Architecture:

2 Inputs->3 Hidden Neurons(Sigmoid)->1 Output (sigmoid)

Output=1->hifh risk op=0->low risk

```python
import numpy as np
np.random.seed(0)
x=np.array([
    [120,85,150,95],   #Glucose
    [25,30,35,22]      #BMI
])
x=x/200

#Normalize Values
x=x/200
#op:1=High risk , 0=Low risk
y=np.array([[1,0,1,0]])

print("x shape:",x.shape)
print("y shape:",y.shape)
```

# Sigmoid Function

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

# Initialize the parameters

```python
#Hidden Layer(3 neurons,2 inputs)
w1=np.random.randn(3,2)*0.1
b1=np.zeros((3,1))

#Output LAyer(1 neuron,3 hidden inputs)
w2=np.random.randn(1,3)*0.1
b2=np.zeros((1,1))
```

# Forward Propagation

```python
#hidden Layer
z1=np.dot(w1,x)+b1
a1=sigmoid(z1)

#Output Layer
z2=np.dot(w2,a1)+b2
a2=sigmoid(z2)

print('Predicted output:')
print(a2)
```

# Loss Function

Binary Cross Entropy

```python
m=y.shape[1]
loss=-np.sum(y*np.log(a2+1e-8)+(1-y)*np.log(1-a2+1e-8))/m
print('Loss:',loss)
```

# Back Propagation

output Layer Gradient

```python
dz2=a2-y
dw2=(1/m)*np.dot(dz2,a1.T)
db2=(1/m)*np.sum(dz2,axis=1,keepdims=True)
```

# Hidden Layer Gradient

Sigmoid derivative:

```python
da1=np.dot(w2.T,dz2)
dz1=da1*a1*(1-a1)
dw1=(1/m)*np.dot(dz1,x.T)
db1=(1/m)*np.sum(dz1,axis=1,keepdims=True)

learning_rate=0.5
w2-=learning_rate*dw2
b2-=learning_rate*db2
w1-=learning_rate*dw1
b1-=learning_rate*db1
```

# Complete Training Loop

```python
epoch=5000
for i in range(epoch):
  #Forward
  z1=np.dot(w1,x)+b1
  a1=sigmoid(z1)


  z2=np.dot(w2,a1)+b2
  a2=sigmoid(z2)

  #loss
  loss=-np.sum(y*np.log(a2+1e-8)+(1-y)*np.log(1-a2+1e-8))/m

  #backward
  dz2=a2-y
  dw2=(1/m)*np.dot(dz2,a1.T)
  db2=(1/m)*np.sum(dz2,axis=1,keepdims=True)

  da1=np.dot(w2.T,dz2)
  dz1=da1*a1*(1-a1)

  dw1=(1/m)*np.dot(dz1,x.T)
  db1=(1/m)*np.sum(dz1,axis=1,keepdims=True)

  #update
  w2-=learning_rate*dw2
  b2-=learning_rate*db2
  w1-=learning_rate*dw1
  b1-=learning_rate*db1

  if i%100==0:
    print('Epoch:',i,'Loss:',loss)


```
