# **Diabetes Risk Prediction(using TensorFlow Framework)**

Neural Network Architecture:

2 Inputs->3 Hidden Neurons(Sigmoid)->1 Output (sigmoid)

Output=1->hifh risk op=0->low risk

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
x=np.array([
   [120,25],
   [85,30],
   [150,35],
   [95,22]
])
x=x/200
y=np.array([1,0,1,0])
print("input shape:",x.shape)
print("output shape:",y.shape)
```

# Build Neural Network Model

```python
model=Sequential([
    Dense(3,input_shape=(2,),activation='sigmoid'),
    Dense(1,activation='sigmoid')
])
model.summary()
```

# Compile the model

```python
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
```

# Train the Model

```python
model.fit(x,y,epochs=500,verbose=0)
```

# Make Predictions

```python
predictions=model.predict(x)
print("predictions:")
print(predictions)
```

# **Iris Dataset**
Goal:
Predict which Flower species an iris belongs to.
1 setosa 2 veriscolor 3 virginica

Input layer(4 neurons) | hidden Layer(10 neurons,reLU) | Hidden Layers(5 neurons,ReLU) | output Layer(3 neurons,Softmax)

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow import argmax
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

```

```python
#Load iris dataset
iris=load_iris()
x=iris.data
y=iris.target
target_names=iris.target_names
print(target_names)
```

```python
#split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#standard scaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#define the model Architeecture
model=Sequential()
model.add(Dense(10,activation='relu',input_shape=(x_train.shape[1],)))
model.add(Dense(5,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train_scaled,y_train,epochs=50,validation_split=0.1,verbose=1)
```

```python
test_loss,test_accuracy=model.evaluate(x_test_scaled,y_test)
print(f'Test Accuracy:',test_accuracy)
```

```python
#prediction
y_pred_proba=model.predict(x_test_scaled)
y_pred=argmax(y_pred_proba,axis=1)#it selects highest probability


#accuracy
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```
