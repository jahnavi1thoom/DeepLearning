# **Perceptron(Binary Classification)**



Example:study hours->pass/fail

```python
import numpy as np
```

**step1:Prepare Training data**

```python
x=np.array([1,2,3,4])  #study hours
y=np.array([0,0,1,1])  #fail(0)/pass(1)
x,y
```

**step2 : initialize parameters**

```python
w=0.0
b=0.0
learning_rate=1
epochs=5
print("initial weight:",w)
print("initial Bias:",b)
```

**Define Activation Function(Step Function)**

```python
def step_function(z):
  return 1 if z>=0 else 0
```

**Training Loop**

```python
for epoch in range(epochs):
  print(f"\n Epoch {epoch+1}")
  for i in range(len(x)):
    #1.compute weighted sum
    z=w*x[i] + b

    #2.apply activation funcn(step func)
    y_pred=step_function(z)

    #3.compute the error
    error=y[i]-y_pred

    #4.update the weight
    w=w+learning_rate*error*x[i]

    #5.update bias
    b=b+learning_rate*error

    print(f'Input:{x[i]},Preedicted:{y_pred},Actual:{y[i]},w:{w},b:{b}')


```

**final Model Parameters**

```python
print("final weight",w)
print("final bias:",b)
```

**Testing the Model**

```python
test_value=2.9
z=w*test_value+b
prediction=step_function(z)
print(f"For study hours {test_value}, Prediction",prediction)
```

```python
#plot with decision boundary
import matplotlib.pyplot as plt

# Create a scatter plot of the training data
plt.scatter(x, y, color='blue', marker='o', label='Training Data')
# Calculate the decision boundary
# The decision boundary is where w*x_boundary + b = 0, so x_boundary = -b/w
if w != 0:
    decision_boundary_x = -b / w
    plt.axvline(x=decision_boundary_x, color='red', linestyle='--', label=f'Decision Boundary (x={decision_boundary_x:.2f})')
else:
    plt.axvline(x=0, color='red', linestyle='--', label='Decision Boundary (w=0)')


# Add labels and title
plt.xlabel('Study Hours (x)')
plt.ylabel('Result (0: Fail, 1: Pass)')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```

# **Perceptron AND Gate**

```python
import numpy as np
```

```python
x=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
    ])
y=np.array([0,0,0,1])
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10
print("Initial weights",w)
print('initial Bias:',b)

```

```python
def step_function(z):
   return 1 if z>=0 else 0
```

```python
#Training loop
for epoch in range(epochs):
  print(f"epoch{epoch+1}")
  for i in range(len(x)):
    #1.compute weighted sum
    z=np.dot(w,x[i])+b

    #2.compute activqtion func
    y_pred=step_function(z)

    #3.compart prediction with actual
    error=y[i]-y_pred

    #4.update weights
    w=w+learning_rate*error*x[i]

    #5.bias
    b=b+learning_rate*error

    print(f"input:{x[i]},predicted:{y_pred},actual:{y[i]},w:{w},b:{b}")
```

```python
print("\n Final weights:",w)
print("final Bias",b)
```

test model

```python
print('\n testing the model AND GAte')
for i in range(len(x)):
  z=np.dot(w,x[i])+b
  prediction=step_function(z)
  print('input ',x[i],"prediction:",prediction)
```

```python
import matplotlib.pyplot as plt

# Create a meshgrid for plotting the decision boundary
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5
h = 0.01  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict output for each point in the meshgrid
Z = np.array([step_function(np.dot(w, [x1, x2]) + b) for x1, x2 in np.c_[xx.ravel(), yy.ravel()]])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

# Plot the training points
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k', marker='o', s=100)

plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('Perceptron Decision Boundary for AND Gate')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks([0, 1])
plt.yticks([0, 1])
plt.grid(True)
plt.show()
```

 Business Scenario: Email Spam Detection for a Startup
A startup company is building a basic email filtering system to classify emails as:
1  Spam
0  Not Spam
The company wants to first implement a simple Perceptron model before moving to complex neural networks.
They identified two key binary features from emails:
Contains Suspicious Keywords (e.g., lottery, prize, urgent)
Yes  1
No  0
Sender is Unknown
Yes  1
No  0

Part 1  Conceptual Understanding
Is this dataset linearly separable?
Can a single-layer Perceptron solve this problem?
What logical function does this resemble?


Part 3  Decision Boundary
After training:
Write the final equation of the decision boundary.
Plot the decision boundary.
Explain geometrically how Perceptron separates spam and non-spam emails.
Part 4  Critical Thinking
What happens if we change E3 output to 0?
Will Perceptron still converge?
If not, why?
What limitation of Perceptron does this expose? **bold text**


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x1 = suspicious keywords
# x2 = unknown sender
# y  = spam or not spam

X = np.array([
    [0,0],  # E1
    [0,1],  # E2
    [1,0],  # E3
    [1,1]   # E4
])
y = np.array([0,0,0,1])  # AND logic
```

```python
w = np.array([0.0, 0.0])   # w1, w2
b = 0.0                    # bias
learning_rate = 1          #  = 1
epochs = 10
```

```python
# Step activation function
def step(z):
    if z >= 0:
        return 1
    else:
        return 0
```

```python
# ==========================================
# PART 3: TRAIN PERCEPTRON
# ==========================================

print("TRAINING STARTED\n")

for epoch in range(epochs):

    print("Epoch", epoch+1)

    error_count = 0

    for i in range(len(X)):

        x = X[i]
        target = y[i]

        # weighted sum
        z = np.dot(w, x) + b

        # predicted output
        y_pred = step(z)

        # error
        error = target - y_pred

        # update weights
        w = w + learning_rate * error * x
        b = b + learning_rate * error

        print("Input:", x,
              "Target:", target,
              "Predicted:", y_pred,
              "Error:", error,
              "Updated weights:", w,
              "Bias:", b)

        if error != 0:
            error_count += 1

    print()

    # stop if converged
    if error_count == 0:
        print("Perceptron converged!")
        break
```

```python
print("\nFINAL MODEL")
print("Weights:", w)
print("Bias:", b)

print("\nDecision Boundary Equation:")
print(f"{w[0]}*x1 + {w[1]}*x2 + {b} = 0")
```

```python
plt.figure()

# plot points
for i in range(len(X)):

    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], marker='o', s=200, label="Not Spam" if i==0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], marker='s', s=200, label="Spam")

# decision boundary
x_vals = np.linspace(-0.5,1.5,100)

if w[1] != 0:
    y_vals = -(w[0]*x_vals + b)/w[1]
    plt.plot(x_vals, y_vals)

plt.title("Decision Boundary - Email Spam Detection")
plt.xlabel("Suspicious Keywords (x1)")
plt.ylabel("Unknown Sender (x2)")
plt.grid()
plt.legend()

plt.show()
```

```python
# ==========================================
# PART 6: GEOMETRIC INTERPRETATION
# ==========================================

print("\nGEOMETRIC INTERPRETATION:")
print("Perceptron creates a straight line separating spam and not spam emails.")
print("Spam emails lie on one side, non-spam on the other.")
```

```python
# ==========================================
# PART 7: CRITICAL THINKING (Non-linear case)
# Change dataset to XOR
# ==========================================

print("\nCRITICAL THINKING CASE (XOR)")

y_xor = np.array([0,1,1,0])

w2 = np.array([0.0,0.0])
b2 = 0.0

for epoch in range(10):

    error_count = 0

    for i in range(len(X)):

        x = X[i]
        target = y_xor[i]

        z = np.dot(w2,x)+b2
        pred = step(z)

        error = target-pred

        w2 = w2 + learning_rate*error*x
        b2 = b2 + learning_rate*error

        if error!=0:
            error_count+=1

    print("Epoch",epoch+1,"Errors:",error_count)

print("\nPerceptron does NOT converge for XOR")
print("This shows limitation: Perceptron only works for linearly separable data")
```
